```python
# imports
import pandas as pd
import numpy as np
import random
```

```python
# This keeps the random data consistent every time the script runs. Without a seed, every run would generate different data, so this makes testing easier.
np.random.seed(3)
random.seed(3)
```

```python
# This is the number of records I want for my dataset.
num_records = 600
```

```python
# Here I'm creating lists of possible values for each category. The script is going to randomly select from these lists when its generating data. 
genders = ['Female', 'Male']
workout_types = ['Strength Training', 'Cardio', 'HIIT', 'Yoga', 'Mixed']
intensity_levels = ['Low', 'Medium', 'High']
hydration_levels = ['Low', 'Medium', 'High']
yes_no = ['Yes', 'No']
```

```python
# Now I'm creating the main dataset dictionary. 
data = {
    "Age": np.random.randint(18, 60, num_records),
    "Gender": np.random.choice(genders, num_records),
    "Workout_Type": np.random.choice(workout_types, num_records),
    "Workouts_Per_Week": np.random.randint(1, 8, num_records),
    "Workout_Duration": np.random.randint(20, 121, num_records),
    "Intensity_Level": np.random.choice(intensity_levels, num_records),
    "Sleep_Hours": np.round(np.random.uniform(4, 10, num_records), 1),
    "Recovery_Days": np.random.randint(0, 5, num_records),
    "Warm_Up_Completed": np.random.choice(yes_no, num_records),
    "Previous_Injury": np.random.choice(yes_no, num_records),
    "Pain_Level_After_Workout": np.random.randint(0, 11, num_records),
    "Hydration_Level": np.random.choice(hydration_levels, num_records)
}
```

```python
# Here I'm turning the dictionary into a dataframe
df = pd.DataFrame(data)
```

```python
# Now I'm creating a function that calculates the injury risk for each person using a point system.
def assign_injury_risk(row):
    risk_score = 0

    if row['Previous_Injury'] == 'Yes':
        risk_score += 2
    
    if row['Pain_Level_After_Workout'] >= 7:
        risk_score += 3
    elif row['Pain_Level_After_Workout'] >= 4:
        risk_score += 1
    
    if row['Sleep_Hours'] < 6:
        risk_score += 2

    if row['Recovery_Days'] <= 1:
        risk_score += 2

    if row['Intensity_Level'] == 'High':
        risk_score += 2
    elif row['Intensity_Level'] == 'Medium':
        risk_score += 1
    
    if row['Warm_Up_Completed'] == 'No':
        risk_score += 1
    
    if row['Hydration_Level'] == 'Low':
        risk_score += 1

    if risk_score >= 7:
        return 'High'
    elif risk_score >= 4:
        return 'Medium'
    else:
        return 'Low'
```

```python
# This means I'm applying the function to every row in the dataset
df['Injury_Risk'] = df.apply(assign_injury_risk, axis=1)
```

```python
# Removing any duplicates just in case
df = df.drop_duplicates()
```

```python
# Now I'm filtering out any impossible values
df = df[
    (df['Age'].between(18, 60)) &
    (df['Workouts_Per_Week'].between(1, 7)) &
    (df['Workout_Duration'].between(20, 120)) &
    (df['Sleep_Hours'].between(0, 12)) &
    (df['Recovery_Days'].between(0, 4)) &
    (df['Pain_Level_After_Workout'].between(0, 10))
]
```

```python
# Now I'm saving the sataset as a CSV and setting my index to False to keep the dataset a bit cleaner by not including the extra column
df.to_csv('gym_injury_data.csv', index=False)
# Now I'm printing out some basic info about the dataset to make sure it looks right.
print('Dataset created successfully!')
print()
print(df.head())
print()
print('Shape:', df.shape)
print()
print('Injury Risk Counts:')
print(df['Injury_Risk'].value_counts())
```
